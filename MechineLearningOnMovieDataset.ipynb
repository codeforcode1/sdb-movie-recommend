<h2> To take the genres, id, keywords, title, overview, revenue, vote_average, cast </h2>

```python
movies.info()
```

```python
movies = movies[['title','genres','id','keywords','overview','revenue','vote_average','cast', 'crew']]
```

```python
movies.head(1)
```

```python
movies = movies.dropna()
```

```python
 movies.iloc[0]['genres']
```

```python
import ast
```

```python
def convert(obj):
    li=[]
    for i in ast.literal_eval(obj):
        li.append(i['name'])
    return li
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies.iloc[0]['keywords']
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
movies.iloc[0]['cast']
```

```python
def castmain(obj):
#     for i in range(3):
    j=0
    li=[]
    for i in ast.literal_eval(obj):
        li.append(i['name'])
        j=j+1
        if(j==3):
            break
    return li 
```

```python
castmain(movies.iloc[0]['cast'])
```

```python
movies['cast'] = movies['cast'].apply(castmain)
```

```python
movies.iloc[0]['crew']
```

```python
def find_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
find_director(movies.iloc[0]['crew'])
```

```python
movies['crew'] = movies['crew'].apply(find_director)
```

```python
movies.info()
```

```python
movies.iloc[0]['overview']
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
def space(obj):
    l=[]
    for i in obj:
#         print(i,type(i))
        l.append(i.replace(' ',''))
    return l
```

```python
def co(obj):
    l=[]
    for i in obj:
#         print(i,type(i))
        l.append(i.replace(' ',','))
    return l
```

```python
movies['cast'] = movies['cast'].apply(space)
movies['keywords'] = movies['keywords'].apply(space)
movies['overview'] = movies['overview'].apply(co)
movies['crew'] = movies['crew'].apply(space)
```

```python
movies.head()
```

```python
movies['all_tags'] = movies['overview'] + movies['keywords'] + movies['cast'] + movies['crew'] + movies['genres'] #movies['vote_average']
```

```python
movies.head()
```

```python
df = movies[['id','title','vote_average','all_tags']]
```

```python
df.head()
```

```python
df['all_tags'] = df['all_tags'].apply(lambda x:' '.join(x))
```

```python
df.head()
```

```python
df['all_tags'] = df['all_tags'].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stemming(obj):
    l=[]
    for i in obj.split():
        l.append(ps.stem(i))
    return ' '.join(l)
```

```python
stemming(df.iloc[0]['all_tags'])
```

```python
df['all_tags'] = df['all_tags'].apply(stemming)
```

```python
df.shape
```

```python
df = df.drop_duplicates(subset=['vote_average','title','id'])
```

```python
df.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
countvec = CountVectorizer(max_features=10000,stop_words='english')
vector = countvec.fit_transform(df['all_tags']).toarray()
```

```python
countvec.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity  = cosine_similarity(vector)
```

```python
similarity
```

```python
type(len(similarity[0]))
```

```python
vote = np.array(df['vote_average'])
vote[0]
```

```python
similarity = np.array(similarity)
similarity
```

```python
n= len(similarity[0])
sim=  np.ones((n,n))
for i in range(n):
    for j in range(n):
        sim[i][j]=similarity[i][j]*vote[i]*vote[j]
```

```python
sim
```

```python
sorted(list( enumerate(sim[1])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def find_index(obj, id):
    recomanded = sorted(list( enumerate(obj[id])),reverse=True,key=lambda x:x[1])[1:6]
    l=[]
    for i in recomanded:
        l.append(i[0])
    return l
```

```python
find_index(sim,1)
```

```python
def sim_movies(title, check = True):
    for i in range(len(sim[0])):
        if df.iloc[i,1].lower()==title.lower():
            k=i
            break
#     index = df[df['title']==title]
    if check==True:
        flag=sim
    else:
        flag=similarity
    similar_movie_index = find_index(flag,k)
    l=[]
    for i in similar_movie_index:
        l.append(df.iloc[i,1])
    return l
```

```python
sim_movies('iron man')
```

```python
sim_movies('iron man',0)
```
